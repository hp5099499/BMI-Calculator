```python

import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
# syntax for 3-D projection
ax = plt.axes(projection ='3d')
# defining all 3 axes
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
# plotting
ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot geeks for geeks')
plt.show()
c=x+y
ax.scatter(x,y,z,c=c)
ax.set_title('3d Scatter plot geeks for geeks')
plt.show()
```

```python
import pandas as pd
data=pd.read_csv('tips.csv')
x=data['day']
y=data['total_bill']
plt.bar(x,y,color="red",edgecolor='yellow',linewidth=2)
plt.title("Tips Dataset")
plt.xlabel("Day")
plt.ylabel("Total bill")
plt.show()
```

```python
import pandas as pd
data=pd.read_csv('tips.csv')
x=data['total_bill']
plt.hist(x,bins=25,color="violet",edgecolor='white',linestyle='--',alpha=0.5)
plt.title("Tips Dataset")
plt.ylabel("Frequency")
plt.xlabel("Total bill")
plt.show()
```

```python
import pandas as pd
data=pd.read_csv('tips.csv')
x=data['day']
y=data['total_bill']
plt.scatter(x,y,c=data['size'],s=data['total_bill'],marker='D',alpha=0.5)
plt.title("Tips Dataset")
plt.xlabel("Day")
plt.ylabel("Total bill")
plt.show()         
```
