# Create Labels for a Plot

```python
plt.plot(x,y)
plt.title('graph')
plt.xlabel('X-AXIS')
plt.ylabel('y-AXIS')
```

# Set Font Properties for Title and Labels

```python
plt.plot(x,y)
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'red','size':15}

plt.title('graph',fontdict=font1)
plt.xlabel('X-AXIS',fontdict=font2)
plt.ylabel('y-AXIS',fontdict=font2)
```

# Position the Title  --loc  (title()),grid()

```python
plt.plot(x,y)
plt.title('graph',loc='right')  #left,right,center
plt.xlabel('X-AXIS')
plt.ylabel('y-AXIS')
```

```python
plt.plot(x,y)
plt.title('graph')  
plt.xlabel('X-AXIS')
plt.ylabel('y-AXIS')
plt.grid(axis='x',color='g',ls='-.',lw=5)     #also apply axis=x,y
```

# compare plots


```python
x=np.array([0,8,56,23])
y=np.array([0,230,56,23])
plt.scatter(x,y,c='hotpink')
x=np.array([23,45,67,89,26])
y=np.array([45,35,56,67,34])
plt.scatter(x,y,c='g')
```

# color each dots

```python
x=np.array([23,45,67,89,26])
y=np.array([45,35,56,67,34])
colors=np.array(['r','g','m','gold','blue'])
plt.scatter(x,y,c=colors)
```

# colormap   ==cmap='virdis'

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='ocean_r')
plt.colorbar()
```

# size==s

```python
s=np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x,y,s=s)
plt.show()
```

# alpha--adjust transparency of dots

```python
plt.scatter(x,y,s=s,alpha=0.5,c='r')
```

# Combine Color Size and Alpha

```python
x=np.random.randint(100,size=(100))
y=np.random.randint(100,size=(100))
colors=np.random.randint(100,size=(100))
sizes=10*np.random.randint(100,size=(100))
plt.scatter(x,y,c=colors,s=sizes,cmap='nipy_spectral',alpha=0.5)
plt.colorbar()
```

# histogram-showing frequency distributions.

```python
x=np.random.normal(170,10,250)            #250-values,mean=170,sd=10
plt.hist(x,color='gold',alpha=0.5)
```

# Pie Charts

```python
y=np.array([35,25,25,15])
mylabels=["S","T","U","V"]
plt.pie(y,labels=mylabels,startangle=90)
```

# explode-- how far from the center each wedge is displayed

```python
m=[0.2,0,0,0]
colors=['r','g','blue','deepskyblue']
plt.pie(y,labels=mylabels,startangle=90,explode=m,shadow=True,colors=colors)
plt.legend(title='Four stages')
```
