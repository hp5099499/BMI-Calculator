```python
print(ps.isnull().sum())
ps.isnull().all()
ps["Years of Experience"].isnull().any()
```

```python
ps['Years of Experience']=ps['Years of Experience'].fillna=('na')
# ps['Years of Experience']
```

```python
ps['Years of Experience']=ps['Years of Experience'].replace(to_replace="na",value='2.0')
ps
ps
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
gp=ps.groupby("Years of Experience",axis=1)
print(gp)
```

```python
ps["Gender"]=ps["Gender"].str.replace("Female","F")
ps["Gender"]=ps["Gender"].str.replace("Male","M")
ps["Gender"]=ps["Gender"].str.replace("Other","O")
```

```python
print(ps.head())
```

```python
plt.figure(figsize=(7,4))
ax=sns.countplot(data=ps,x="Years of Experience")
plt.bar_label(ax.containers[0])
plt.show()
```

```python
x=ps["Salary"]
y=ps["Years of Experience"]
fig = plt.figure(figsize=(7,7), facecolor='c', edgecolor='g',linewidth=2)
plt.scatter(x,y)
plt.title('Salary vs Years Of Experience',fontsize=25 ,color="blue")
plt.xlabel("Salary")
plt.ylabel("Years of Experience")
plt.ylim(0,25)
plt.show()
```

```python
x=ps['Gender']
y=ps["Salary"]
plt.bar(x,y,color='green',edgecolor='yellow',linewidth=2)
plt.title("Line Chart")
plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")
plt.show()
```

```python
x=ps['Years of Experience']
plt.hist(x,bins=25,color="violet",edgecolor='white',linestyle='--',alpha=0.5)
plt.title("Tips Dataset")
plt.ylabel("Frequency")
plt.xlabel("Total bill")
plt.show()
```

```python
x=ps["Gender"]
y=ps["Salary"]
plt.plot(x,y,yamarker="D",linewidth=2,linestyle="dotted")
plt.show()
```
