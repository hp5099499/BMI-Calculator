# Linear Regression, Logistic Regression, Decision Tree

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("emp.csv")
df.head()
```

```python
x=df.drop('Salary',axis='columns')
y=df['Salary']
```

```python
from sklearn.preprocessing import LabelEncoder
Company_1=LabelEncoder()
Job_1=LabelEncoder()
Degree_1=LabelEncoder()
```

```python
x['Company_1']=Company_1.fit_transform(x['Company'])
```

```python
x['Job_1']=Job_1.fit_transform(x['Job'])
```

```python
x
```

```python
x['Degree_E']=Degree_1.fit_transform(x['Degree'])
```

```python
x
```

```python
y
```

```python
y.replace({'<50000':0,'>50000':1})
```

```python
x1 = x.drop(['Company','Job','Degree',"Job_1"],axis='columns')
```

```python
x1.head(5)
```

```python
# ml model
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x1,y)
```

```python
model.score(x1,y)
```

```python
model.predict([[2,1,0]])
```

```python
df= pd.read_csv("dia.csv")
df.sample()
```

```python
x=df.drop('Outcome',axis='columns')
x.sample()
```

```python
features_col=['Pregnancies',	'Glucose',	'BloodPressure',	'SkinThickness','BMI','DiabetesPedigreeFunction','Age']
```

```python
y=df['Outcome']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
576/768
```

```python
# model
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
predicted_value=model.predict(x_test)
```

```python
y_test
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test,predicted_value)
```
