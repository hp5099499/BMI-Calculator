```python
import matplotlib.pyplot as plt
df.plot(y="Kilometers_Driven",x="Price",kind="scatter")
plt.xlim(xmin=0)
plt.ylim(ymin=0)
plt.show()
```

```python
import numpy as mp
import seaborn as sns
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df=df.drop("S.no.",axis=1)
df.head()
```

```python
# change  
df['Fuel_Type']=df['Fuel_Type'].str.replace('Petrol',"CNG")
df.head()
```

```python
plt.figure(figsize=(4,4))
a=sns.countplot(data=df,x="Fuel_Type")
a.bar_label(a.containers[0])
plt.title("Fuel_Type Distribution")
plt.show()
```

```python
# fuel type comparison
g=df.groupby("Transmission").agg({"Kilometers_Driven":'mean'})
print(g)
```

```python
sns.heatmap(g,annot=True)
```

```python
gf=df.groupby("Fuel_Type").agg({"Kilometers_Driven":'mean','Price':"mean"})
print(gf)
```

```python
sns.heatmap(gf,annot=True)
```

```python
go=df.groupby("Owner_Type").agg({"Kilometers_Driven":'mean'})
print(go)
```

```python
sns.heatmap(go,annot=True)
```

```python
import seaborn as sns

sns.lineplot(x='Fuel_Type',y="Kilometers_Driven",data=df,marker='o',hue="Year",style="Year",palette="Accent",legend="full",dashes=False)
plt.grid()
plt.title("Distance-Fuel Graph According to Respestive Year",c="b")
plt.show()
```

```python
sns.barplot(x='Fuel_Type',y="Kilometers_Driven",data=df,hue="Year",ci=20,orient="v",saturation=50,errcolor="b",errwidth="1.7",
           capsize=0.01,alpha=0.9)
sns.set(style="darkgrid")
plt.show()
```
