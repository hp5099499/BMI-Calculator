```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins=sns.load_dataset('penguins')
penguins.sample(3)
```

```python
sns.jointplot(x='bill_length_mm',y='bill_depth_mm',data=penguins)
```

```python
sns.pointplot(x='bill_length_mm',y='bill_depth_mm',data=penguins)
```

```python
a=np.random.rand(10,12)
sns.heatmap(a,annot=True)
```

```python
sns.violinplot(x='bill_length_mm',y='sex',data=penguins)
```

```python
sns.pairplot(penguins)
```

```python
sns.PairGrid(penguins)
```

```python
sns.stripplot(x='bill_length_mm',y='bill_depth_mm',data=penguins)
```

```python
tips=sns.load_dataset('tips')
tips.sample(3)
```

```python
sns.boxplot(x='total_bill',y='day',data=tips)
```

```python
sns.boxenplot(x='',y='day',data=tips)
```
